```python
X = np.zeros((200,350*350))
bad_indices = []
for i, val in enumerate(glob.glob('images/*.jpg')):
    try:
        img = mpimg.imread(val)
        img = img/255.
        if(img.shape[0]==350 and img.shape[1]==550):
            X[i]=img.reshape((np.prod(img.shape[:2])))
        else:
            bad_indices.append(i)
    except:
        pass
```

```python
def processData(labels, images):
    print('Processing data')
    labels_array = np.array(list(map(int, labels)))
    images_array = np.zeros(shape=(len(images), 48, 48))
    for i, row in enumerate(images):
        image = np.fromstring(row, dtype=int, sep=' ')
        image = np.reshape(image, (48, 48))
        images_array[i] = image
        if i%1000 == 0:
            print('.', sep='\t')
    print('\nDone pre-processing')
    return labels_array, images_array
```
